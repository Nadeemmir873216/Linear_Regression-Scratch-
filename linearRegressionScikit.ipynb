## Simple Linear Regression with Machine Learning Library

```python
import pandas as pd, numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#sample data
x = np.array([4,2,3,4,5, 78, 56, 34, 23, 54, 34, 23, 45 ,65, 23, 34, 5,56 ,7,54 ]).reshape(-1,1)
y = np.array([2,3,2,4,5, 5,23, 45, 67, 23, 76, 23, 7, 34, 23, 56, 56, 76, 7,8])
```

```python
# create and train model
model = LinearRegression()
model.fit(x,y)
```

```python
y_pred = model.predict(x)
```

```python
#printing coef
print("slope (m)", model.coef_[0])
print("intercept (b)", model.intercept_)
```

```python
# plotting result
plt.scatter(x,y, color='blue', label = 'Actual data')
plt.plot(x, y_pred, color='red', label = 'regression line')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid (True)
plt.show()
```

## Simple Linear Regression without Machine Learning Library

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= x.flatten()
print(x)
print(y)
```

```python
#Mean
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
# (m) & (b)
numerator = sum((x-x_mean)*(y-y_mean))
denominator = sum((x-x_mean)**2)
```

```python
m = numerator/denominator
b = y_mean -m*x_mean
```

```python
y_pred = m* x +b
print(y_pred)
```

```python
print("Slope (m) ", m)
print("Intercept (b) ",b)
```

```python
# plotting result
plt.scatter(x,y, color='blue', label = 'Actual data')
plt.plot(x, y_pred, color='red', label = 'regression line')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid (True)
plt.show()
```
